# Python Übung 1 - Daten untersuchen

```python
data1 = [[10, 8.04], [8, 6.95], [13, 7.58], [9, 8.81], [11, 8.33], [14, 9.96], [6, 7.24], [4, 4.26], [12, 10.84], [7, 4.82], [5, 5.68]]
data2 = [[10, 9.14], [8, 8.14], [13, 8.74], [9, 8.77], [11, 9.26], [14, 8.1], [6, 6.13], [4, 3.1], [12, 9.13],[7, 7.26], [5, 4.74]]
data3 = [[10, 7.46], [8, 6.77], [13, 12.74], [9, 7.11], [11, 7.81], [14, 8.84], [6, 6.08], [4, 5.39], [12, 8.15], [7, 6.42], [5, 5.73]]
data4 = [[8, 6.58], [8, 5.76], [8, 7.71], [8, 8.84], [8, 8.47], [8, 7.04], [8, 5.25], [19, 12.5], [8, 5.56], [8, 7.91], [8, 6.89]]
```

1: Benutzen Sie eine Python-Funktion, um die Anzahl der Dimensionen p zu berechnen für alle 4 Datenreihen. 
Hinweis: Importieren Sie pandas und erstellen Sie 4 DataFrames für die Daten, benennen Sie die Spalten X und Y.

```python
import pandas as pd

dataFrame1 = pd.DataFrame(data=data1, columns=["X", "Y"])
dataFrame2 = pd.DataFrame(data=data2, columns=["X", "Y"])
dataFrame3 = pd.DataFrame(data=data3, columns=["X", "Y"])
dataFrame4 = pd.DataFrame(data=data4, columns=["X", "Y"])

# display(dataFrame1, dataFrame2, dataFrame3, dataFrame4)
print("Datenstruktur: (N, p)")
print("1:", dataFrame1.shape)
print("2:", dataFrame2.shape)
print("3:", dataFrame3.shape)
print("4:", dataFrame4.shape)
```

2: Geben Sie die Anzahl der Datenpunkte, den Durchschnitt, Min, Max, und die Perzentilen aus. Vergleichen Sie die Werte über alle Datensätze.

```python
print("Eigenschaften von 1:")
print(dataFrame1.describe())
print()
print("Eigenschaften von 2:")
print(dataFrame2.describe())
print()
print("Eigenschaften von 3:")
print(dataFrame3.describe())
print()
print("Eigenschaften von 4:")
print(dataFrame4.describe())
```

3: Berechnen Sie die Korrelation zwischen X und Y für alle 4 Datensätze. Was können Sie daraus schließen? Berechnen Sie zudem auch die Korrletion ^2. 

```python
corr1 = dataFrame1.corr().iloc[1][0]
print("Korrelation bei 1: normal / quadriert")
print(corr1, corr1 ** 2)
print()

corr2 = dataFrame2.corr().iloc[1][0]
print("Korrelation bei 2: normal / quadriert")
print(corr2, corr2 ** 2)
print()

corr3 = dataFrame3.corr().iloc[1][0]
print("Korrelation bei 3: normal / quadriert")
print(corr3, corr3 ** 2)
print()

corr4 = dataFrame4.corr().iloc[1][0]
print("Korrelation bei 4: normal / quadriert")
print(corr4, corr4 ** 2)
```

4: Erstellen Sie Scatterplots für alle vier Datensätze. Hint: Nutzen Sie Seaborn oder Matplotlib. 

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(40, 10))
plt.subplot(1, 4, 1)
sns.scatterplot(x="X", y="Y", data=dataFrame1)
plt.title("DataFrame1")

plt.subplot(1, 4, 2)
sns.scatterplot(x="X", y="Y", data=dataFrame2)
plt.title("DataFrame2")

plt.subplot(1, 4, 3)
sns.scatterplot(x="X", y="Y", data=dataFrame3)
plt.title("DataFrame3")

plt.subplot(1, 4, 4)
sns.scatterplot(x="X", y="Y", data=dataFrame4)
plt.title("DataFrame4")

plt.show()
```

5: Führen Sie eine lineare Regression durch. Der unten gegebene Code importiert bereits numpy und das Modell als auch den Plot. Das reshaping für die Lineare Regression ist auch schon erledigt. 

```python
#import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

X1 = dataFrame1.iloc[:, 0].values.reshape(-1, 1)
Y1 = dataFrame1.iloc[:, 1].values.reshape(-1, 1)
reg1 = LinearRegression()
reg1.fit(X1, Y1)
score1 = reg1.score(X1, Y1)

X2 = dataFrame2.iloc[:, 0].values.reshape(-1, 1)
Y2 = dataFrame2.iloc[:, 1].values.reshape(-1, 1)
reg2 = LinearRegression().fit(X2, Y2)
score2 = reg2.score(X2, Y2)

X3 = dataFrame3.iloc[:, 0].values.reshape(-1, 1)
Y3 = dataFrame3.iloc[:, 1].values.reshape(-1, 1)
reg3 = LinearRegression().fit(X3, Y3)
score3 = reg3.score(X3, Y3)

X4 = dataFrame4.iloc[:, 0].values.reshape(-1, 1)
Y4 = dataFrame4.iloc[:, 1].values.reshape(-1, 1)
reg4 = LinearRegression().fit(X4, Y4)
score4 = reg4.score(X4, Y4)
```

5: contd. Danach müssen Sie das Objekt LinearRegression instanziieren und "fitten". Sie können die Regression mit score bewerten. Geben Sie die Koeffizienten aus (coef_) und die Schnittstelle (intercept_). Vergleichen Sie den score (R\^2-Wert) mit dem oben berechneten Korrelation ^ 2 Wert.

```python
print("Lineare Regression: Bewertung (Korrelation^2), Koeffizient, Schnittstelle")
print(f"1: {score1} ({corr1 ** 2}), {reg1.coef_[0][0]}, {reg1.intercept_[0]}")
print(f"2: {score2} ({corr2 ** 2}), {reg2.coef_[0][0]}, {reg2.intercept_[0]}")
print(f"3: {score3} ({corr3 ** 2}), {reg3.coef_[0][0]}, {reg3.intercept_[0]}")
print(f"4: {score4} ({corr4 ** 2}), {reg4.coef_[0][0]}, {reg4.intercept_[0]}")
```

6: Sagen die die Werte für Y_pred mit predict voraus. Nutzen Sie einen Scatterplot, um X, Y und die Regressionslinie einzuzeichnen. Das geht z. B. mit Matplotlib oder Seaborn. 

```python
plt.figure(figsize=(40, 10))

# Jeweils immer prediction erstellen, dann scatter mit Ursprungsdaten, dann Linie durch Nutzung von y_predict

y_predict1 = reg1.predict(X1)
plt.subplot(1, 4, 1)
plt.scatter(X1, Y1)
plt.plot(X1, y_predict1)
plt.title("Reg1")

y_predict2 = reg2.predict(X2)
plt.subplot(1, 4, 2)
plt.scatter(X2, Y2)
plt.plot(X2, y_predict2)
plt.title("Reg2")

y_predict3 = reg3.predict(X3)
plt.subplot(1, 4, 3)
plt.scatter(X3, Y3)
plt.plot(X3, y_predict3)
plt.title("Reg3")

y_predict4 = reg4.predict(X4)
plt.subplot(1, 4, 4)
plt.scatter(X4, Y4)
plt.plot(X4, y_predict4)
plt.title("Reg4")

plt.show()
```
